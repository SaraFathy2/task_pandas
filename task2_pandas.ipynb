```python
def clean_residence(res):
    if pd.isna(res):
        return res
    
    res = res.strip().lower()
    
    mapping = {
        'bi residence': 'BI Residence',
        'bi-residence': 'BI Residence',
        'biresidence': 'BI Residence',
        'bi_residence': 'BI Residence',
        'sognsvann': 'Sognsvann',
        'private': 'Private'
    }
    
    return mapping.get(res, res.title())
```

```python
df_data['residence'] = df_data['residence'].apply(clean_residence)
```

```python
df_data["residence"].value_counts()
```

```python
def clean_country(country):
    if pd.isna(country):
        return country
    
    country = country.strip().lower()
    
    mapping = {
        'norway': 'Norway',
        'norge': 'Norway',
        'rsa': 'South Africa'
    }
    
    return mapping.get(country, country.title())
```

```python
df_data['country'] = df_data['country'].apply(clean_country)
```

```python
df_data["country"].value_counts()
```

```python
df_data["gender"].value_counts()
```

```python
def clean_gender(gender):
    if pd.isna(gender):
        return gender
    
    gender = gender.strip().lower()
    
    mapping = {
        'f': 'Female',
        'm': 'Male',
        'female': 'Female',
        'male': 'Male'
    }
    
    return mapping.get(gender, gender.title())

```

```python
df_data['gender'] = df_data['gender'].apply(clean_gender)
```

```python
def clean_education(edu):
    if pd.isna(edu):
        return edu
    
    edu = edu.strip().lower()
    
    mapping = {
        'highschool': 'High School',
        'high school': 'High School',
        'diploma': 'Diploma',
        'diplomaaa': 'Diploma',
        'diplomaa': 'Diploma',
        'bachelors': 'Bachelors',
        'barrrchelors': 'Bachelors',
        'masters': 'Masters',
        'doctorate': 'Doctorate'
    }
    
    return mapping.get(edu, edu.title())
```

```python
df_data['prevEducation'] = df_data['prevEducation'].apply(clean_education)
```
